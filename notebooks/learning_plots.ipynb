```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

Bar Charts

```python
# group by genre and calculate Ave total streams

avg_streams = spotify_df.groupby('Genre')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.figure(figsize=(10,6))
colors = ["#756bb1"]*len(avg_artist_streams)

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#18df28"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

special_artist = "Ariana Grande"
colors = ["#df1893" if artist == special_artist else "gray" for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.9)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Ariana Grande ranks 2nd in Total Average Streams")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

Scatter Plots

```python
spotify_df.columns
spotify_df.info()
```

```python
plt.scatter(spotify_df["Streams Last 30 Days (Millions)"], spotify_df["Total Streams (Millions)"])

plt.show()
```

```python
plt.Figure(figsize=(10,6))

plt.scatter(spotify_df["Streams Last 30 Days (Millions)"], 
            spotify_df["Total Streams (Millions)"],
            color = "#0ce74e",
            alpha = 0.8,
            s = 25)

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation between Month Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.show()
```

```python
plt.Figure(figsize=(10,6))

featured_genre = "Pop"

colors = ["#0ce74e" if genre == featured_genre else "#d1d5d1" for genre in spotify_df["Genre"]]

plt.scatter(spotify_df["Streams Last 30 Days (Millions)"], 
            spotify_df["Total Streams (Millions)"],
            color = colors,
            alpha = 0.8,
            s = 25)

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18, pad = 25)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)

plt.show()
```

Line Charts

```python
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(10,6))
plt.plot(
    yearly_avg.index,
    yearly_avg.values,
    marker="+",
    linestyle="--",
    color="#0ce74e",
    linewidth = 0.5)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream Less in 2022?", fontsize=18, pad=20)


ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(
    2022,
    color="#0e6e2b",
    linewidth = 0.5,
    alpha = 0.7
    )

ax.text(
    2022+0.05,
    2225,
    "Dip in Total Streams",
    color = "#0e6e2b",
    fontsize = 8
    )

plt.ylim(0, 3500)

plt.show()
```

Pie Charts 

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels=None,
    startangle=110,
    colors=["gray", "#1faf4b"],
    autopct="%1.1f%%",
    )

plt.title("Near Even Split Between Premium and Free Users")

plt.legend(
    labels = platform_totals.index,
    loc = "lower right",
    title = "Platform"
)

plt.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

wedges, exts, autotexts = ax.pie(
    platform_totals,
    autopct="%1.1f%%",
    startangle=110,
    colors=["gray", "#1faf4b"]
)

ax.set_title(
    "Near Even Split Between Premium and Free Users",
    fontsize=30,
    pad=15,
    loc="center"
    )


ax.legend(
    wedges,
    platform_totals.index,
    loc="lower center",
    bbox_to_anchor = (0.5, 0.95),
    ncol = 2
)

for autotext in autotexts:
    autotext.set_fontsize(30)
    autotext.set_color("white")
    autotext.set_weight("bold")


plt.tight_layout()

plt.show()
```
